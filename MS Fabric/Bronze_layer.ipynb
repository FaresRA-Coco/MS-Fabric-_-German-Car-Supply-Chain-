```python
df = spark.read.table("bronze_CarSupplyChainGermanCar")
display(df)
```

```python
df.describe()
```

```python
from pyspark.sql import functions as F
df = df.withColumn(
    "time_stamp",
    F.to_timestamp(F.col("timestamp"),
    "yyyy-MM-dd'T'HH:mm:ss.SSSSSS")
)
```

```python
df = df.drop("timestamp")
```

```python
display(df)
df.describe
```

```python
df.write.mode("overwrite").option("header","true").format("csv").save("Files/GermanCar")
```
